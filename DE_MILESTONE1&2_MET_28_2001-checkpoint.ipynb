## Handling Missing data

Assuming all missing values are MCAR 

```python
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_colwidth', None)
```

```python
for column in df_accidents.columns:
    print(column)
    print("-1 VALUES :")
    print(df_accidents[column].iloc[np.where(df_accidents[column] == "-1")].value_counts().values)
    print("DATA missing VALUES :")
    print(df_accidents[column].iloc[np.where(df_accidents[column] == "Data missing or out of range")].value_counts().values)
    print("---------------------------------------------")
```

```python
def data_cleaning_df(df):
    df_cleaned = df.copy()
    null_values_threshold = 0.2
    # These 2 columns represent an exception and needs to be handled before cleaning the data
    df_cleaned.second_road_number.replace('first_road_class is C or Unclassified. These roads do not have official numbers so recorded as zero ' , 0.0 , inplace=True)
    df_cleaned.first_road_number.replace('first_road_class is C or Unclassified. These roads do not have official numbers so recorded as zero ' , 0.0 , inplace=True)
    
    
    for column in df.columns:
        data_missing_count = df[column].iloc[np.where(df[column] == "Data missing or out of range")].value_counts().values
        negative_one_count = df[column].iloc[np.where(df[column] == "-1")].value_counts().values
        if(data_missing_count.size > 0):
            null_value_percentage = (data_missing_count[0]/df[column].shape[0])*100
            if(null_value_percentage < null_values_threshold):
                df_cleaned[column].replace("Data missing or out of range" , np.nan , inplace=True)
            else:
                mode_value = df[column].value_counts().index[0]
                df_cleaned[column].replace("Data missing or out of range" , mode_value , inplace=True)
        if(negative_one_count.size > 0):
            null_value_percentage = (negative_one_count[0]/df[column].shape[0])*100
            if(null_value_percentage < null_values_threshold):
                df_cleaned[column].replace("-1" , np.nan , inplace=True)
            else:
                mode_value = df[column].value_counts().index[0]
                df_cleaned[column].replace("-1" , mode_value , inplace=True)
    
    # Columns of type object, needs to be converted to float
    df_cleaned.first_road_number = df_cleaned.first_road_number.astype(float)
    df_cleaned.second_road_number = df_cleaned.second_road_number.astype(float)
    
    df_cleaned_final = df_cleaned.dropna()
    return df_cleaned_final
```

By replacing `Unclassified` values in both the first_road_number and second_road_number with the float value `0.0`, we removed all custom null values and introduced a new float value to provide consistency with the rest of the attribute's values along with enabling more complex analysis

The above function `data_cleaning_df` takes a dataframe and cleans its data from all the custom and pre-defined `null` values. It does this by first checking for those null values and counts their frequency in a specific column. If it exceeds a pre-defined threshold (by comparing the ratio of the null value compared to the whole column) , we use mode imputation method and replace every null value with the mode , while not exceeding it will result in replacing those custom null values with `NaN` value. This is to ensure that most of the data is present and not greatly modified even after removing all the `NaN` null values from the dataset

```python
df_clean = data_cleaning_df(df_accidents)
for column in df_clean.columns:
    print(column)
    print("-1 VALUES :")
    print(df_clean[column].iloc[np.where(df_clean[column] == "-1")].value_counts().values)
    print("DATA missing VALUES :")
    print(df_clean[column].iloc[np.where(df_clean[column] == "Data missing or out of range")].value_counts().values)
    print("---------------------------------------------")
```

After removing the remaining null values using the `dropna` method, we have now successfully cleaned the data from all null values and we can move on into transforming our new dataset to fit our analytical needs

Getting multivariante outliers of number_of_vehicles and number_of_casualties attributes, where we calculate the lof of them, in which we get the value = -1 to represent an outlier and value = 1 which indicates a non-outlier. According this idea, we have y-pred array which rpresent the value of l where l is the localoutlier factor. So, using this array to trace back and get the original point in array X that have the original values of number_of_vehicles and number_of_casualties, so the format is as follows [[number_of_vehicles number_of_casualties]....]. At this momment we got the values that caused to have the outliers and their reference index, therefore using the function dataset.column_name.iloc[index]=new value, in oreder to access each row by index and specifying the column name you want to change the value and replace it with the new value. In our case, we calculated the median of each column "number_of_vehicles" and "number_of_casualties", then replaced the outliers with these medians. At last, we had a check before and after detecting outliers and replacing them.

```python
def label_encoding_df(df):
    label_encoding_list = ['day_of_week','road_type','junction_detail','junction_control','pedestrian_crossing_physical_facilities',
                          'light_conditions','weather_conditions','road_surface_conditions','special_conditions_at_site','carriageway_hazards',
                          'police_force','local_authority_district','local_authority_ons_district','local_authority_highway']
    extra_columns = ['date','time','lsoa_of_accident_location','accident_year','accident_reference','first_road_number','second_road_number']
    total_columns = label_encoding_list + extra_columns
    result = df.copy() 
    for column in result.columns:
        if result.dtypes[column] == object and total_columns.count(column) > 0:
        
            result[column] = preprocessing.LabelEncoder().fit_transform(result[column])
    
    onehot_encoding_df = result.copy()

    # Select the numeric columns
    numeric_subset = result.select_dtypes('number')
    # Select the categorical columns
    categorical_subset = result.select_dtypes('object')
    
    # Creating a dummy df to host the categorical columns while excluding the label encoded columns + extra_columns 
    new_categorical_subset = pd.get_dummies(categorical_subset)
    new_categorical_subset_k1 = pd.get_dummies(categorical_subset,drop_first=True, dummy_na=True)
    
    onehot_encoding_df = pd.concat([numeric_subset, new_categorical_subset], axis = 1)
    return onehot_encoding_df
```

```python
def label_encoding_df_generic(df):
    unique_threshold = 5
    result = df.copy() 
    #removing the date,time,and lsoa of accident location due to them not needing to be encoded
    new = result.filter(['date','time','lsoa_of_accident_location'], axis=1)
    result.drop(['date','time','lsoa_of_accident_location'], axis=1 , inplace=True)
    
    cumulative_mapping = {}
    for column in result.columns:
        if result.dtypes[column] == object and df[column].unique().size > unique_threshold:
        
            le = preprocessing.LabelEncoder()
            mapped = le.fit_transform(result[column])

            mapping = dict(zip(le.classes_, range(0, len(le.classes_)+1)))
            #print(mapping) prints encodings of values
            cumulative_mapping.update({column:mapping})
            result[column] = mapped
    
    # EXporting the encodings to a csv file
    csv_df = pd.DataFrame(cumulative_mapping)
    
    csv_df.to_csv('file.csv')
    
    onehot_encoding_df = result.copy()

    # Select the numeric columns
    numeric_subset = result.select_dtypes('number')
    # Select the categorical columns
    categorical_subset = result.select_dtypes('object')
    
    # Creating a dummy df to host the categorical columns while excluding the label encoded columns + extra_columns 
    new_categorical_subset = pd.get_dummies(categorical_subset)
    new_categorical_subset_k1 = pd.get_dummies(categorical_subset,drop_first=True, dummy_na=True)
    
    onehot_encoding_df = pd.concat([numeric_subset, new_categorical_subset], axis = 1)
    return onehot_encoding_df
```

The above function `label_encoding_df_generic` takes a dataframe and encodes the appropriate columns to either one-hot encoding or label encoding. The function decides the appropriate encoding technique by using a `unique_threshold` variable, any column containing higher number of unique values than the variable will be encoded using label encoding, while the rest of the categorical columns will be one-hot encoded. This is done to prevent from one-hot encoding a column with too many unique values that could result in too many newly created columns, while also using the one-hot encoding on columns with few unique values in order to avoid the ordering issue that could result from the label encoding technique

## Scaling

Scaling is used to make all features contribute the same amount in prediction.After studing the ranges of our attibutes, so for attributes of large ranges that could outweigh variables with smaller ranges and will distort the distance computations for these methods. 

so for attributes with a range more than 30 we would scale it, most of the attributes will be scaled by standard scaler of Z score when we don't know for sure there is a maximum value and for attributes like speed limit (like age ) there is a maximum so we will use max min scaling 

Regarding the location (geo information) such as Latitude and Longitude won't be scaled since they are values should not be altered drastically that is why it should not be scaled to avoid the changing the meaning of values as well as date and time

```python
res["local_authority_district"].min()
```

```python
def Normalize(df):
    result = df.copy()    
    for column in result:
        
        maxValue = result[column].max();
        minValue = result[column].min();
        rangeV = maxValue - minValue
        
        values = result[column].unique()
        if (rangeV<30 or np.array_equal(values, [0,1]) or np.array_equal(values, [1,0] or column =="accident_year" or column =="accident_reference" or column =="day_of_week" or column == "location_easting_osgr" or column == "location_northing_osgr" or column == "longitude" or column == "latitude" )):
            continue
#       elif(column == "longitude" or column == "latitude"):
#             x = np.cos(result["latitude"]) * np.cos(result["longitude"])
#             y = np.cos(result["latitude"]) * np.sin(result["longitude"])
#             z = np.sin(result["latitude"]) 
#             result['longitude'] = x
#             result['latitude'] = y
        elif(column == "speed_limit" or column == "number_of_casualties" or column == "police_force"):
            #min-max
            newCol = MinMaxScaler().fit_transform(result[[column]]) 
            result[column] = newCol
        else:
            #Z-score 
            newCol = StandardScaler().fit_transform(result[[column]])
            result[column] = newCol
    return result
```
